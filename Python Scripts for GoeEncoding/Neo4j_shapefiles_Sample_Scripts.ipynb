```python
import json 
import csv
```

```python
# append properties json values into list
g_properties = []
for item in geo_json['features']:
    g_properties.append(item['properties'])
g_properties
```

```python
# cross check for data integrity
len(g_properties)
```

```python
# append coordinates values into list
g_wkt = []
for item in geo_json['features']:
    g_wkt.append(item['geometry'])
```

```python
# transform the gemetry coordinates into WKT format
g_wkt_final = [convert.geojson_to_wkt(item) for item in g_wkt]
```

Below: write properties list of json to csv, write geomtry list of josn to csv, and concat them together.


```python
#with open('data.json') as json_file: 
#    data = json.load(json_file) 
  
#geojson_data = geo_json['features']
  
# now we will open a file for writing 
data_file = open('data_file.csv', 'w') 
  
# create the csv writer object 
csv_writer = csv.writer(data_file) 
  
# Counter variable used for writing  
# headers to the CSV file 
count = 0
  
for item in g_properties: 
    if count == 0: 
  
        # Writing headers of CSV file 
        header = item.keys() 
        csv_writer.writerow(header) 
        count += 1
  
    # Writing data of CSV file 
    csv_writer.writerow(item.values()) 
  
data_file.close() 
```
